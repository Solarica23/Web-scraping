```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.flipkart.com/samsung-galaxy-f41-fusion-blue-64-gb/p/itmcdc5e93b44d55?pid=MOBFV5PWHRYTYYJF&lid=LSTMOBFV5PWHRYTYYJFLE0GZO&marketplace=FLIPKART&srno=s_1_1&otracker=search&fm=organic&iid=602bc018-3f90-4cf9-9aed-32f4e0234554.MOBFV5PWHRYTYYJF.SEARCH&ssid=te27mbz6dc0000001609137107744&qH=fe546279a62683de"
```

```python
r= requests.get(url)
htmlContent= r.content
print(htmlContent)
```

```python
soup= BeautifulSoup(htmlContent,'lxml')
print(soup.prettify())
```

```python
desc= soup.find('div',id='container')
productDesc=desc.find('div',class_="_2k6Cpt").text
print(productDesc)
```
